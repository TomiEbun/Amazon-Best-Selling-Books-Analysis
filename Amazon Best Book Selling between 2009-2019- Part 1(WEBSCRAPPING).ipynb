```python
!pip install webdriver-manager
!pip install --upgrade selenium

import numpy as np
import pandas as pd

import requests
from selenium import webdriver
from bs4 import BeautifulSoup
from urllib.request import urlopen
from time import sleep

from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

# To bypass warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
### Data Gathering
### Getting urls and the second page for each year beginning from 2009 to 2019)
urls = []
years = [str(i) for i in range(2009, 2020)] 

for year in years:
    urls.append(f"https://www.amazon.com/gp/bestsellers/{year}/books")
    urls.append(f"https://www.amazon.com/gp/bestsellers/{year}/books/?ie=UTF8&pg=2")

urls
```

```python
###Creating a function to get the details for each book and year from (2009-2019)
def get_dir(book,year): 
    '''to get the details of each book for each year''' 
    
    import numpy as np
    '''to get the name of price'''
    try:
        title = book.find('div',class_="_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y").text
    except Exception as e:
        title = np.nan
    try:
        author = book.find('a',class_="a-size-small a-link-child").text 
    except Exception as e:
        author = np.nan  
    try:
        cover_type = book.find('span',class_="a-size-small a-color-secondary a-text-normal").text
    except Exception as e:
        cover_type = np.nan
    try:
        ratings= book.find('span',class_="a-icon-alt").text[:3] 
    except Exception as e:
        ratings = np.nan 
    try:
        ranks = book.find('span', class_='zg-bdg-text').text[1:]
    except Exception as e:
        ranks = np.nan   
    try:
        price = book.find('span',class_="_cDEzb_p13n-sc-price_3mJ9Z").text[1:]
    except Exception as e:
        price = np.nan
    try:
        no_of_reviews  = book.find('span',class_="a-size-small").text
    except Exception as e:
        no_of_reviews = np.nan
    year = year
    return [title,author,cover_type,ratings,ranks,price,no_of_reviews,year]
```

```python
##Creating list that contains the set of each year
year = [(str(i),str(i)) for i in range(2009,2020)] 
years = [j for i in year for j in i] 
years
```

```python
##To get the books in every pages- first and second of each year from  (2009- 2019)
urls = ['https://www.amazon.com/gp/bestsellers/2009/books',
 'https://www.amazon.com/gp/bestsellers/2009/books/?ie=UTF8&pg=2',
 'https://www.amazon.com/gp/bestsellers/2010/books',
 'https://www.amazon.com/gp/bestsellers/2010/books/?ie=UTF8&pg=2',
 'https://www.amazon.com/gp/bestsellers/2011/books',
 'https://www.amazon.com/gp/bestsellers/2011/books/?ie=UTF8&pg=2',
 'https://www.amazon.com/gp/bestsellers/2012/books',
 'https://www.amazon.com/gp/bestsellers/2012/books/?ie=UTF8&pg=2',
 'https://www.amazon.com/gp/bestsellers/2013/books',
 'https://www.amazon.com/gp/bestsellers/2013/books/?ie=UTF8&pg=2',
 'https://www.amazon.com/gp/bestsellers/2014/books',
 'https://www.amazon.com/gp/bestsellers/2014/books/?ie=UTF8&pg=2',
 'https://www.amazon.com/gp/bestsellers/2015/books',
 'https://www.amazon.com/gp/bestsellers/2015/books/?ie=UTF8&pg=2',
 'https://www.amazon.com/gp/bestsellers/2016/books',
 'https://www.amazon.com/gp/bestsellers/2016/books/?ie=UTF8&pg=2',
 'https://www.amazon.com/gp/bestsellers/2017/books',
 'https://www.amazon.com/gp/bestsellers/2017/books/?ie=UTF8&pg=2',
 'https://www.amazon.com/gp/bestsellers/2018/books',
 'https://www.amazon.com/gp/bestsellers/2018/books/?ie=UTF8&pg=2',
 'https://www.amazon.com/gp/bestsellers/2019/books',
 'https://www.amazon.com/gp/bestsellers/2019/books/?ie=UTF8&pg=2']

### Setting up ChromeOptions to disable the proxy server
options = webdriver.ChromeOptions()
options.add_argument("--no-proxy-server")

### Initializing the Chrome driver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

all_books = []

try:
    for url in urls:
        driver.get(url)

        # Adding a sleep mechanisms to ensure that the page loads completely before scraping
        time.sleep(5)  

        the_soup = BeautifulSoup(driver.page_source, 'html.parser')
        books = the_soup.find_all(id = 'gridItemRoot')                 

        # Appending the books for each year to the 'all_books' list
        all_books.append(books)

finally:
    ### Quit the driver after scraping all the URLs
    driver.quit()
```

```python
len(all_books), len(years)
```

```python
## Getting index for the year  for easier looping through
year_index = (list(enumerate(years)))
yc = year_index
```

```python
books_data= [] 
for i in yc:   
    for books in all_books[i[0]]:             
        for book in books:                  
            books_data.append(get_dir(book,i[1]))  
            
books_data 
```

```python
#### open file
with open('Amazon.txt', 'w+') as f:
     
    # Elements of list
    for items in books_data: 
        try:
            f.write('%s\n' %items)
        except Exception as e:
            f.write('%s\n' 'nothing')
     
    print("File written successfully")

f.close()
```

```python
### Converting the books data extracted to a dataframe
amazon_best_selling_books= pd.DataFrame(books_data, columns = [
                         'title',
                         'author',
                         'cover_type',
                         'ratings',
                         'ranks',
                         'price',
                         'no_of_reviews',
                         'year'])
```
